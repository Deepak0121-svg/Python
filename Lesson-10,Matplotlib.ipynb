```python
from matplotlib import pyplot as plt
fig=plt.figure()
axes=fig.add_axes([0.1,0.7,0.6,0.8])#control the left,right,width,height of the canvas from(0to2)
axes.plot(x,y,'b--')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
for ax in axes:
    ax.plot(x,y,'b*')
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)
for ax in axes:
    ax.plot(x,y,'b')
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.7,0.5,0.6])#big axes
axes2=fig.add_axes([0.5,0.4,0.6,0.8])#small canvas
axes1.plot(x,y,'orange')
axes2.plot(y,x,'black')

```

```python
#New graphic object
fig=plt.figure(figsize=(19,5),dpi=7)
fig.add_subplot()
plt.plot(x,y,'orange')
```

```python
#set title,xandylabel to axes
fig,axes=plt.subplots()
axes.set_title("Curve")
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.plot(x,y,'orange')
```

```python
fig,axes=plt.subplots(dpi=200)
axes.plot(x,x+2,color='black',linewidth=0.25)
axes.plot(x,x+4,color='red',linewidth=1)
line,=axes.plot(x,x+6,color='blue',lw=1.25)
line.set_dashes([5,10,3,15])
line,=axes.plot(x,x+4,color='black',lw=1.25)
line.set_dashes([5,10,15,15])
```

```python
#set marker

fig,ax=plt.subplots()
ax.plot(x,x+1,color='black',marker='^',markerfacecolor='white')
ax.plot(x,x+2,color='blue',marker='*',markersize=10)
ax.plot(x,x+3,color='yellow',marker='o',markerfacecolor='red')
```

```python
# setting line style
fig,ax=plt.subplots(dpi=130)
ax.plot(x,x+1,color='black',linestyle='-.',marker='^',markerfacecolor='yellow')
ax.plot(x,x+2,color='red',linestyle=':')
ax.plot(x,x+7,color='orange',linestyle='dotted')
```

```python
'''set canvas grid and axis'''
fig,axes=plt.subplots(1,2,figsize=(12,6))
axes[0].plot(x,x**2,x,x**3,lw=2,linestyle=':') 
axes[0].grid(True)

axes[1].plot(x,x**2,x,x**3,lw=2,linestyle=':',marker='^',markerfacecolor='red',markersize=8)
axes[1].set_ylim([0,50])
axes[1].set_xlim([2,8])
axes[1].grid(True)
```

# other 2D Graphics

```python
n=np.array([0,1,2,3,4])
fig,axes=plt.subplots(1,4,figsize=(19,4))
axes[0].set_title("scatter")
axes[0].scatter(n,n**2,color="violet",marker='^')

axes[1].set_title("bar")
axes[1].bar(n,n**3,align="center",alpha=.5,color="orange")

axes[2].set_title("step")
axes[2].step(n,n**2,lw=2,linestyle='-.',color='black')

axes[3].set_title("fill_between")
axes[3].fill_between(n,n**4,color="yellow")
```

```python
'''Draw a radar chart'''
fig=plt.figure(figsize=(7,20))
ax=fig.add_axes([0.0,0.2,0.7,1.0],polar=True)
E=np.linspace(0,2*np.pi,100)
ax.plot(E,E*3,color='#0ef0e4',lw=2)
```

```python
''' Draw a histogram '''
n=np.array([0,1,2,3,4,8])
fig,axes=plt.subplots(1,3,figsize=(10,3))
axes[0].set_title("Histogram")
axes[0].hist(n,color="red",lw=2)

axes[1].set_title("cumulative_histogram")
axes[1].hist(n,cumulative=True,bins=250)

axes[2].set_title("Fillbetween")
axes[2].fill_between(n,n*3,color="#07fa64")
```
